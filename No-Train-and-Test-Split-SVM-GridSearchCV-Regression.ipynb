```python
import pandas as pd
```

```python
dataset=pd.read_csv("Insurance_pre.csv")
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
dataset=dataset.replace({True:0,False:1})
```

```python
dataset
```

```python
dataset.columns
```

```python
independent=dataset[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
dependent=dataset[['charges']]
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
independent=sc.fit_transform(independent)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVR
param_grid={'kernel':['linear','poly','sigmoid','rbf'],
            'C':[10,100,1000,2000,3000], 'gamma':['scale','auto']}
regressor=GridSearchCV(SVR(),param_grid,refit=True,verbose=3,n_jobs=-1)
regressor.fit(independent,dependent)
```

```python
result=regressor.cv_results_
table=pd.DataFrame.from_dict(result)
table
```

```python
print("The r_score values best parameter {}".format(regressor.best_params_))
```

```python
Age=float(input("Age ="))
BMI=float(input("BMI ="))
Children=input("Children =")
Sex=int(input("Sex 0 or 1 ="))
Smoke=int(input("Smoke 0 or 1="))
```

```python
Future_Pred=regressor.predict([[Age,BMI,Children,Sex,Smoke]])
print("Future Prediction = {}".format(Future_Pred))
```
