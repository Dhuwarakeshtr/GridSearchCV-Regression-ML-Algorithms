```python
import pandas as pd
```

```python
dataset=pd.read_csv("insurance_pre.csv")
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
dataset=dataset.replace({True:0,False:1})
```

```python
dataset.columns
```

```python
independent=dataset[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
dependent=dataset[['charges']]
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor
param_grid={'criterion':['squared_error', 'friedman_mse', 'absolute_error', 'poisson'], 
            'max_features':['sqrt', 'log2',1.0], 'n_estimators':[10,100]}
regressor=GridSearchCV(RandomForestRegressor(),param_grid,refit=True,verbose=3,n_jobs=-1)
regressor.fit(independent,dependent)
```

```python
result=regressor.cv_results_
table=pd.DataFrame.from_dict(result)
table
```

```python
print("The r_score value for best parameter {}".format(regressor.best_params_))
```

```python
Age=float(input("Age ="))
BMI=float(input("BMI ="))
Children=input("Children =")
Sex=int(input("Sex 0 or 1 ="))
Smoke=int(input("Smoke 0 or 1 ="))
```

```python
future_pred=regressor.predict([[Age,BMI,Children,Sex,Smoke]])
print("Future Prediction {}".format(future_pred))
```
